# 求一个集合的所有k大小的子集

```python
import math
import csv
def ziji(items,k):
    sub_list_all = []
    N = len(items)
    for i in range(2 ** N):
        combo = []
        for j in range(N):
            # test jth bit of integer i
            if (i >> j) % 2 == 1:
                combo.append(items[j])
        if(len(combo)==k):
            sub_list_all.append(combo)
    return sub_list_all
```
